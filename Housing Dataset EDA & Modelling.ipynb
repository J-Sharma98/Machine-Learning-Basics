## Building Linear Regression Model to predict Sale Price without the categorical variables

### In this section, we will build a model and exclude all the categorical variables.

```python
# Seprating the dependent and independent variables.

x = df_num.drop('SalePrice',axis=1)
y = df_num['SalePrice']
x.head()
```

```python
# Split the data into x_train and x_test, y_train and y_test

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50)
```

```python
# looking at the shape of train and test data.

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import statsmodels.api as sm
```

```python
# Building linear regression model 1

lin_reg_model_1 = sm.OLS(y_train, x_train).fit()
lin_reg_model_1.summary()
```

###### In this model, we can see that the Adjusted R Sqaured value is 0.975. Considering the fact that we have excluded the categorical variables and still the model is performing so well, we can assume that the model is overfitting. It can also be seen that there are many variables which are not significant (that is having a p-value of less than 0.05). We can exclude those variables from our model. 

```python
# Pass x_test to the model to predict y.

predictions_model_1 = lin_reg_model_1.predict(x_test)
predictions_model_1
```

```python
# Compare predicted y and y_test using RMSE (Root Mean Square Error)
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, predictions_model_1)
rmse = np.sqrt(mse)
rmse

# We can see that the RMSE of this model is very large. It is to be observed that even though the model has a very high
# adjusted R squared value, the error is very large. It means that our model is performing badly.
```

## Building Linear Regression Model to predict Sale Price with the categorical variables (Dummy Encoding)

#### Including the categorical variables and encoding them, we will create a new model.

```python
# Encoding dummy variables

df_cat_encoded = pd.get_dummies(df_cat, drop_first=True)
df_cat_encoded.head()
```

```python
# Normalization

for column in df_num:
    df_num[column] = (df_num[column] -
                           df_num[column].mean()) / df_num[column].std()  
```

```python
# Combing the encoded categorical variables and normalised numeric variables. 

df_combine = pd.concat([df_cat_encoded,df_num], axis = 1)
```

```python
df_combine.head(2)
```

```python
# Seprating the dependent and independent variables.

x = df_combine.drop('SalePrice',axis=1)
y = df_combine['SalePrice']
x.head()
```

```python
# Split the data into x_train and x_test, y_train and y_test

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50)
```

```python
# Looking at the shape of train and test data

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Building the second regression model.

lin_reg_model_2 = sm.OLS(y_train, x_train).fit()
lin_reg_model_2.summary()
```

###### It can be seen that the Adjusted R Squared of this model is 0.916, which is less than the previous model. But we also need to look at the RMSE of this model to accurately judge which model is better. In this model as well there are many features which are not significant. 

```python
# Pass x_test to the model to predict y.

predictions_model_2 = lin_reg_model_2.predict(x_test)
predictions_model_2
```

```python
# Looking at the RMSE of this model

mse = mean_squared_error(y_test, predictions_model_2)
rmse = np.sqrt(mse)
rmse

# We can see that the RMSE of this model is significantly lower than the previous model. The RMSE of this model is extremely
# low, which means that this model is better than the previous model.
```
